```python
import cv2
import os
from tqdm import tqdm

REAL_VIDEOS = "/content/drive/MyDrive/celeb-df/Celeb-real"
FAKE_VIDEOS = "/content/drive/MyDrive/celeb-df/Celeb-synthesis"

DATASET_PATH = "/content/drive/MyDrive/deepfake_dataset"
REAL_IMAGES = os.path.join(DATASET_PATH, "real")
FAKE_IMAGES = os.path.join(DATASET_PATH, "fake")

os.makedirs(REAL_IMAGES, exist_ok=True)
os.makedirs(FAKE_IMAGES, exist_ok=True)

def extract_frames(video_dir, save_dir, frame_skip=30):
    for root, _, files in os.walk(video_dir):
        for file in tqdm(files):
            if not file.endswith(".mp4"):
                continue
            cap = cv2.VideoCapture(os.path.join(root, file))
            frame_id = 0
            while True:
                ret, frame = cap.read()
                if not ret:
                    break
                if frame_id % frame_skip == 0:
                    name = f"{file}_{frame_id}.jpg"
                    cv2.imwrite(os.path.join(save_dir, name), frame)
                frame_id += 1
            cap.release()

print("Extracting REAL frames...")
extract_frames(REAL_VIDEOS, REAL_IMAGES)

print("Extracting FAKE frames...")
extract_frames(FAKE_VIDEOS, FAKE_IMAGES)

print("✅ Frame extraction complete")
```
